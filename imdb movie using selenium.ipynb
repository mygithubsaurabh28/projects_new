```python
import pandas as pd
import numpy as np
import requests
import selenium 
from selenium import webdriver
from time import sleep
from random import randint
pd.set_option("display.max_columns",None)
```

```python
driver=webdriver.Chrome("E:\chromedriver.exe")
```

```python
name_movie=[]
year_of_release=[]
ratings=[]
```

```python
pages=np.arange(1,1000,100)
```

```python
for page in pages:
    page=driver.get("https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start="+str(page)+"&ref_=adv_nxt" )
    movie_name=driver.find_elements_by_xpath('//h3[@class="lister-item-header"]/a')
    sleep(randint(2,8))
    year_movie=driver.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']")
    for i in movie_name:
        name_movie.append(i.text)
        rating_movie=driver.find_elements_by_xpath("//div[@class='inline-block ratings-imdb-rating']")
        year_of_release.append(i.text)
        ratings.append(i.text)
```
